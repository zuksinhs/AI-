```python
class Rectangle:
  def __init__(self, name, width, height):
    self.name = name
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

  def perimeter(self):
    return 2 * self.width + 2 * self.height

class Square(Rectangle):
  def __init__(self, name, length):
    super().__init__(name, length, length)
```

```python
rectangle1 = Rectangle("rectangle1", 2, 3)
rectangle2 = Rectangle("rectangle2", 7, 4)
square1 = Square("square1", 8)

for item in [rectangle1, rectangle2, square1]:
  print(f"{item.name}の面積: {item.area()}、周長: {item.perimeter()}")
```
