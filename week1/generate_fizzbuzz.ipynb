```python
def generate_fizzbuzz(max_num):
  list1=[]
  for i in range(1,max_num+1):
    if i % 15 == 0:
      list1.append("FizzBuzz")
    elif i % 5 == 0:
      list1.append("Buzz")
    elif i % 3 == 0:
      list1.append("Fizz")
    else:
      list1.append(i)
  return list1
```

```python
print(generate_fizzbuzz(100))
```
